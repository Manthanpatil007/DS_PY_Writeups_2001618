 c. Propose 3-4 Analytical questions about the data and show answers to them

1)Is there a significant difference in BMI between patients with benign vs. malignant breast cancer?

```python
import pandas as pd
from scipy.stats import ttest_ind

data = pd.read_csv('/content/My_Drive/MyDrive/Classroom/dataR2.csv')

benign_bmi = data[data['Classification'] == 0]['BMI']
malignant_bmi = data[data['Classification'] == 1]['BMI']
t, p = ttest_ind(benign_bmi, malignant_bmi, equal_var=False)

print(f"t = {t:.2f}, p = {p:.4f}")
```

3)Do the levels of adiponectin and leptin differ significantly between patients with BMI < 30 and BMI >= 30?

```python
import pandas as pd
from scipy.stats import ttest_ind


data = pd.read_csv('/content/My_Drive/MyDrive/Classroom/dataR2.csv')

low_bmi_adiponectin = data[data['BMI'] < 30]['Adiponectin']
high_bmi_adiponectin = data[data['BMI'] >= 30]['Adiponectin']
t1, p1 = ttest_ind(low_bmi_adiponectin, high_bmi_adiponectin, equal_var=False)

low_bmi_leptin = data[data['BMI'] < 30]['Leptin']
high_bmi_leptin = data[data['BMI'] >= 30]['Leptin']
t2, p2 = ttest_ind(low_bmi_leptin, high_bmi_leptin, equal_var=False)


print(f"Adiponectin: t = {t1:.2f}, p = {p1:.4f}")
print(f"Leptin: t = {t2:.2f}, p = {p2:.4f}")
```

The Breast Cancer Coimbra dataset contains various clinical features of patients diagnosed with breast cancer, along with the target variable indicating whether the cancer is benign or malignant. Based on this data, we can define a binary classification problem in which we aim to predict the target variable (i.e., whether the cancer is benign or malignant) based on the clinical features of the patients.

Here are two distinct ML algorithm-based solutions to this problem:

Support Vector Machines (SVM): SVM is a popular algorithm for binary classification problems. It works by finding the hyperplane that maximizes the margin between the two classes. In this case, we can use SVM to learn a decision boundary that separates benign and malignant cancer cases based on the clinical features of the patients. We can experiment with different kernels (such as linear, polynomial, and radial basis function) to find the one that works best for this dataset.

Random Forest: Random Forest is an ensemble learning algorithm that combines multiple decision trees to make predictions. It is particularly useful when dealing with high-dimensional datasets with many features. In this case, we can use Random Forest to build a set of decision trees, each trained on a different subset of the features, and then combine their predictions to obtain a final prediction. We can experiment with different hyperparameters (such as the number of trees and the maximum depth of each tree) to optimize the performance of the model.

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

data = pd.read_csv('/content/My_Drive/MyDrive/Classroom/dataR2.csv')

X_train, X_test, y_train, y_test = train_test_split(data.drop('Classification', axis=1), 
                                                    data['Classification'], 
                                                    test_size=0.2, 
                                                    random_state=42)


svm_clf = SVC(kernel='rbf')
svm_clf.fit(X_train, y_train)
svm_pred = svm_clf.predict(X_test)
svm_acc = accuracy_score(y_test, svm_pred)

print('SVM Classifier Accuracy:', svm_acc)

rf_clf = RandomForestClassifier(n_estimators=100, max_depth=10)
rf_clf.fit(X_train, y_train)
rf_pred = rf_clf.predict(X_test)
rf_acc = accuracy_score(y_test, rf_pred)

print('Random Forest Classifier Accuracy:', rf_acc)
```
